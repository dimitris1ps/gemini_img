```python
model = genai.GenerativeModel(model_name="models/gemini-1.5-flash")
```

```python

response = model.generate_content(
    ["What is the model, brand, and year of this car? The response should be in the form of a json: {brand: ,model:, year:} just provide the json nothing else", img]
)
```
